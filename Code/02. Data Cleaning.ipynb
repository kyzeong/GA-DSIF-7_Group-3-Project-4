```python
# concat weather1 and weather2
weather = pd.concat([weather1,weather2])
weather.head()
```

```python
# drop null as it is less than 5% of whole dataset
weather.dropna(inplace=True)
```

```python
weather.isnull().sum()
```

```python
weather.apply(lambda col: col.str.contains('-'), axis=1).sum()
```

```python
# Replace '-' with 0
def float_checker(row):
    try:
        res = float(row)
    except:
        res = 0
    return res
```

```python
c = ['Daily Rainfall Total (mm)',\
             'Highest 30-min Rainfall (mm)',\
             'Highest 60-min Rainfall (mm)',\
             'Highest 120-min Rainfall (mm)',\
             'Mean Temperature (°C)',\
             'Maximum Temperature (°C)',\
             'Minimum Temperature (°C)',\
             'Mean Wind Speed (km/h)',\
             'Max Wind Speed (km/h)']
```

```python
for col in c:
    weather[col] = weather[col].map(float_checker)
```

```python
weather.info()
```

```python
weather['Date'] = weather['Date']+' '+weather['year'].astype(str)
```

```python
weather["Date"]= pd.to_datetime(weather["Date"])
```

```python
# change column name by replacing ' (mm)' with ''
weather.columns = weather.columns.str.replace(' (mm)', '', regex=False)
```

```python
# change column name by replacing ' (°C)' with ''
weather.columns = weather.columns.str.replace(' (°C)', '', regex=False)
```

```python
# change column name by replacing ' (km/h)' with ''
weather.columns = weather.columns.str.replace(' (km/h)', '', regex=False)
```

```python
# change column name by replacing '-' with '_'
weather.columns = weather.columns.str.replace('-', '_', regex=False)
```

```python
# change column name to lowercase and replace ' ' with '_'
weather.columns = weather.columns.str.replace(' ', '_', regex=False).str.lower()
```

**Create 14 rolling average for each region**

```python
region_names = dict()
for k, v in weather.groupby('region'):
    region_names[k] = v
```

```python
region_names['Admiralty']
```

```python
def rolling_mean(df):
    df['daily_rainfall_total_14'] = df['daily_rainfall_total'].rolling(14).mean()
    df['highest_30_min_rainfall_14'] = df['highest_30_min_rainfall'].rolling(14).mean()
    df['highest_60_min_rainfall_14'] = df['highest_60_min_rainfall'].rolling(14).mean()
    df['highest_120_min_rainfall_14'] = df['highest_120_min_rainfall'].rolling(14).mean()
    df['mean_temperature_14'] = df['mean_temperature'].rolling(14).mean()
    df['maximum_temperature_14'] = df['maximum_temperature'].rolling(14).mean()
    df['minimum_temperature_14'] = df['minimum_temperature'].rolling(14).mean()
    df['mean_wind_speed_14'] = df['mean_wind_speed'].rolling(14).mean()
    df['max_wind_speed_14'] = df['max_wind_speed'].rolling(14).mean()
    return df
```

```python
for r in region_names.values():
    r.sort_values('date', ascending = True, inplace=True)
    rolling_mean(r)
```

```python
region_names['Admiralty']
```

```python
region_list = []
for i in region_names.keys():
    temp_df = region_names[i]
    region_list.append(temp_df)

weather = pd.concat(region_list)
```
